## Estadísticas descriptivas

### Subtask:
Calcularemos estadísticas descriptivas para las columnas numéricas (como promedio, desviación estándar, valores mínimo y máximo, etc.) y para las columnas categóricas (como el conteo de valores únicos y el valor más frecuente). Esto nos ayudará a entender la distribución y variabilidad de los datos.

## Manejo de valores faltantes

### Subtask:
Identificaremos si hay valores faltantes en el conjunto de datos y decidiremos cómo manejarlos (por ejemplo, imputarlos o eliminarlos), si es necesario.


**Reasoning**:
Calcular y mostrar el conteo y el porcentaje de valores faltantes para cada columna para identificar las columnas con datos incompletos y la magnitud del problema.



```python
# Calcular el número total de valores faltantes en cada columna
missing_values_count = df.isnull().sum()

# Calcular el porcentaje de valores faltantes en cada columna
missing_values_percentage = (df.isnull().sum() / len(df)) * 100

# Crear un DataFrame para mostrar el conteo y el porcentaje de valores faltantes
missing_values_table = pd.DataFrame({
    'Conteo de Valores Faltantes': missing_values_count,
    'Porcentaje de Valores Faltantes (%)': missing_values_percentage
})

# Mostrar la tabla de valores faltantes
print("Conteo y porcentaje de valores faltantes por columna:")
display(missing_values_table)
```

## Análisis de columnas categóricas

### Subtask:
Exploraremos las columnas categóricas para entender la distribución de las diferentes categorías y si hay alguna categoría que predomine.


**Reasoning**:
Select the object type columns and calculate the number of unique values and their frequency for each.



```python
# Seleccionar columnas de tipo 'object'
categorical_cols = df.select_dtypes(include='object').columns

print("Análisis de Columnas Categóricas:\n")

# Iterar sobre cada columna categórica
for col in categorical_cols:
    print(f"Columna: {col}")

    # Calcular el número de valores únicos
    unique_count = df[col].nunique()
    print(f"Número de valores únicos: {unique_count}")

    # Mostrar el conteo de cada categoría (los más frecuentes)
    print("Conteo de categorías (más frecuentes):")
    display(df[col].value_counts().head(10)) # Mostrar los 10 más frecuentes

    print("-" * 30)
```

## Análisis de columnas numéricas

### Subtask:
Analizaremos las columnas numéricas para entender su distribución, identificar posibles valores atípicos (outliers) y visualizar su comportamiento.


**Reasoning**:
Identificar las columnas numéricas, calcular estadísticas descriptivas adicionales como skewness y kurtosis, y visualizar su distribución usando histogramas para entender su comportamiento y posibles valores atípicos.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Identificar las columnas numéricas
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Columnas numéricas identificadas:")
print(numeric_cols)

print("\nAnálisis de Columnas Numéricas:\n")

# 2. Para cada columna numérica, calcular e imprimir estadísticas descriptivas adicionales
for col in numeric_cols:
    print(f"Columna: {col}")

    # Calcular estadísticas descriptivas básicas (ya se hizo en describe, pero la mostramos de nuevo si es relevante)
    # print("Estadísticas descriptivas:")
    # display(df[col].describe())

    # Calcular skewness y kurtosis
    skewness = df[col].skew()
    kurtosis = df[col].kurtosis()
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis: {kurtosis:.4f}")

    # 3. Crear visualizaciones apropiadas (Histogramas y Box Plots)
    plt.figure(figsize=(12, 5))

    # Histograma
    plt.subplot(1, 2, 1)
    sns.histplot(df[col].dropna(), kde=True) # Usar dropna para manejar valores faltantes si existen
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')

    # Box Plot para identificar outliers
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col].dropna()) # Usar dropna para manejar valores faltantes si existen
    plt.title(f'Box Plot de {col}')
    plt.xlabel(col)

    plt.tight_layout()
    plt.show()

    print("-" * 30)

# 4. Análisis de las visualizaciones (Este paso es interpretativo y se realizará después de ver las gráficas)
# Se observarán las formas de las distribuciones (normal, sesgada, multimodal), la presencia de picos (kurtosis),
# y la existencia de puntos extremos o atípicos en los box plots.
```

## Análisis de relaciones entre variables

### Subtask:
Investigaremos la relación entre diferentes columnas para descubrir patrones o correlaciones interesantes. Esto podría incluir la relación entre el valor de la tarifa y el tráfico, por ejemplo.


**Reasoning**:
Calculate the correlation matrix for the numeric columns to understand the linear relationships between them, as requested by the first two steps of the subtask.



```python
# 1. Calcular la matriz de correlación para las columnas numéricas
# Seleccionar solo las columnas numéricas
numeric_cols = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_cols.corr()

# 2. Mostrar la matriz de correlación
print("Matriz de Correlación de las columnas numéricas:")
display(correlation_matrix)
```

**Reasoning**:
Based on the correlation matrix and the subtask instructions, I will select interesting pairs of variables to visualize with scatter plots and also visualize the relationship between a numeric and a categorical variable using a box plot.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 3. Seleccionar algunas parejas de variables numéricas interesantes para visualizar
# Interesantes correlaciones de la matriz: Trafico vs TraficoExentos787 (0.700272), Lon vs IdPeaje (0.434131)
# También la solicitada: ValorTarifa vs Trafico (-0.127609)

selected_numeric_pairs = [
    ('ValorTarifa', 'Trafico'),
    ('Trafico', 'TraficoExentos787'),
    ('Lon', 'IdPeaje')
]

# 4. Crear gráficos de dispersión (scatter plots) para visualizar la relación entre estas parejas
print("\nGráficos de Dispersión para Parejas de Variables Numéricas:")
for x_col, y_col in selected_numeric_pairs:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x=x_col, y=y_col, alpha=0.6) # Usar alpha para manejar la sobreposición si hay muchos puntos
    plt.title(f'Relación entre {x_col} y {y_col}')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# 5. Análisis visual de los gráficos de dispersión (Este paso es interpretativo y se realizará después de ver las gráficas)
# Se observará si hay patrones lineales o no lineales, la dirección (positiva/negativa) y la fuerza aparente.
# Se prestará atención a la dispersión de los puntos y la presencia de outliers.

# 6. Considerar también la relación entre una columna numérica y una categórica relevante
# Usaremos ValorTarifa (numérica) y IdCategoriaTarifa (categórica) como ejemplo

print("\nAnálisis de Relación entre ValorTarifa e IdCategoriaTarifa:")

# Para evitar que el box plot sea ilegible si hay muchas categorías, seleccionaremos las categorías más frecuentes
# Contar la frecuencia de cada categoría
category_counts = df['IdCategoriaTarifa'].value_counts()
# Seleccionar las N categorías más frecuentes (ej. 10 o 15)
top_n_categories = category_counts.head(15).index.tolist()

# Filtrar el DataFrame para incluir solo las filas con estas categorías
df_filtered_categories = df[df['IdCategoriaTarifa'].isin(top_n_categories)].copy()

# Crear un box plot
plt.figure(figsize=(14, 7)) # Ajustar tamaño para que quepan más categorías
sns.boxplot(data=df_filtered_categories, x='IdCategoriaTarifa', y='ValorTarifa')
plt.title('Distribución de ValorTarifa por IdCategoriaTarifa (Top 15)')
plt.xlabel('Categoría de Tarifa')
plt.ylabel('Valor de Tarifa')
plt.xticks(rotation=45, ha='right') # Rotar etiquetas del eje X para mejor legibilidad
plt.tight_layout() # Ajustar el layout para evitar solapamiento
plt.show()

# 7. Análisis visual del gráfico de caja (Este paso es interpretativo y se realizará después de ver la gráfica)
# Se observará la mediana, la dispersión (rango intercuartílico), y la presencia de outliers para cada categoría.
# Esto indicará cómo varía el valor de la tarifa entre las diferentes categorías.
```
