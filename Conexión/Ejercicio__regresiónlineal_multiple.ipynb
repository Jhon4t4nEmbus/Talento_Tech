```python
# Imporar las librerias y crear el Data frame

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
data = {
    "x": [5000,5570,4350,7900,6800,5400,6900,3900,4200,5780],
    "y": [160000,189380,139200,260700,217600,183600,234600,136500,138600,202300]}

df = pd.DataFrame(data)
df
```

```python
df['xy']= df['x']* df['y']
df
```

```python
df['x^2']=df['x']**2
df
```

```python
# Crear la fila con las sumas
suma = df.sum()
# Aadir al final del DataFrame
df.loc["Total"] = suma

df
```

```python
# Obtener los valores de la fila 'Total'
sum_x = df.loc["Total", "x"]
sum_y = df.loc["Total", "y"]
sum_xy = df.loc["Total", "xy"]
sum_x_cuadrada = df.loc["Total", "x^2"]

# Obtener el nmero de puntos de datos (excluyendo la fila 'Total')
n = len(df) - 1

# Calcular la pendiente (m) usando la frmula de regresin lineal
# m = (n * (xy) - x * y) / (n * (x^2) - (x)^2)
numerator = (n * sum_xy) - (sum_x * sum_y)
denominator = (n * sum_x_cuadrada) - (sum_x**2)

m = numerator / denominator
print(f"La pendiente calculada es: {m}")
```

```python
b = sum_y / n - m * (sum_x / n)
print(f"La interseccin calculada es: {b}")
```

```python
# Eliminar la fila total
if 'Total' in df.index:
  df_plot = df.drop('Total')
else:
  df_plot = df.copy()


# Crear un diagrama de dispersin
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_plot, x='x', y='y')

# Calcular la regresin lineal
slope, intercept, r_value, p_value, std_err = stats.linregress(df_plot['x'], df_plot['y'])
# Utilice los valores x originales para el grfico de la lnea de regresin
x_line = df_plot['x']
y_line = intercept + slope * x_line

#Agregar lnea de regresin: crea un rango de valores x para la lnea segn los datos originales
x_range = np.array([df_plot['x'].min(), df_plot['x'].max()])
y_range = intercept + slope * x_range

plt.plot(x_range, y_range, color='red', label=f'Linear Regression: y = {slope:.2f}x + {intercept:.2f}')


plt.title('Diagrama de dispersin de x vs y con regresin lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
