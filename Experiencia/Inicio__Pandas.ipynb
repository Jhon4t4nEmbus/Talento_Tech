```python
import numpy as np
import pandas as pd
import random

# List of common Latin American names
latin_american_names = [
    "Mateo", "Sofía", "Santiago", "Valentina", "Sebastián", "Isabella", "Nicolás", "Camila",
    "Alejandro", "Valeria", "Diego", "Mariana", "Gabriel", "Luciana", "Andrés", "Gabriela",
    "David", "Victoria", "Daniel", "Martina", "Pablo", "Antonella", "Adrián", "Regina",
    "Javier", "Renata", "Manuel", "Natalia", "Miguel", "Carolina", "José", "Paula",
    "Ricardo", "Emily", "Fernando", "Andrea", "Carlos", "Florencia", "Luis", "Daniela",
    "Jorge", "María Fernanda", "Roberto", "Alejandra", "Eduardo", "Adriana", "Francisco", "Jimena",
    "Sergio", "Angie", "Antonio", "Laura", "Ernesto", "Sara", "Federico", "Diana",
    "Christian", "Nicole", "Gustavo", "Emilia", "Enrique", "Valery", "Hugo", "Ximena",
    "Ignacio", "Amanda", "Joaquín", "Catalina", "Raúl", "Josefina", "Ángel", "Montserrat",
    "Marcos", "Samantha", "Óscar", "Constanza", "Ramón", "Ashley", "Salvador", "Michelle",
    "Alfredo", "Sophia", "Mauricio", "Alexa", "Arturo", "Guadalupe"
]

# Generate 80 random names from the Latin American names list
nombres_adicionales_80_new = random.sample(latin_american_names, 80)

# Generate 80 random weights (between 40 and 100 kg)
pesos_adicionales_80_new = np.random.uniform(40, 100, 80).tolist()

# Generate 80 random heights (between 1.40 and 1.90 m)
alturas_adicionales_80_new = np.random.uniform(1.40, 1.90, 80).tolist()

# Create a DataFrame with the additional 80 data points
diccionario_datos_80_new = {"Nombre": nombres_adicionales_80_new, "Peso": pesos_adicionales_80_new, "Altura": alturas_adicionales_80_new}
df_adicional_new = pd.DataFrame(diccionario_datos_80_new)

# Concatenate the original DataFrame with the new DataFrame
df = pd.concat([df, df_adicional_new], ignore_index=True)

# Calculate BMI (Peso / Altura^2)
df['IMC'] = df['Peso'] / (df['Altura']**2)

# Function to classify based on BMI
def classify_bmi(bmi):
    if bmi < 18.5:
        return 'Bajo peso'
    elif bmi < 25:
        return 'Peso normal'
    elif bmi < 30:
        return 'Sobrepeso'
    else:
        return 'Obesidad' # Added Obesity category

# Apply the classification function to create the 'Clasificación' column
df['Clasificación'] = df['IMC'].apply(classify_bmi)

display(df.tail()) # Display the last few rows to show the added data and new columns
```
